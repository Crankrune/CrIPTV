```python
import json
import re
from pathlib import Path
```

```python
with open("./data/iptv_database.json", "r", encoding="utf-8") as f:
    iptv_database = json.load(f)
```

```python
moveonjoy_count = 0
for channel in iptv_database:
    name = channel["name"]
    url = channel["url"]
    if re.search(r"https?://fl1.moveonjoy", url, flags=re.I):
        moveonjoy_count += 1
        print(f"{name}: {url}")

print(f"moveonjoy count: {moveonjoy_count}")
```
